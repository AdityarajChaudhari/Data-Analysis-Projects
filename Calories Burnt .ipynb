```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
st="‪D:/Datasets/exercise.csv" 
train_data = pd.read_csv(st.strip("‪u202a"))
```

```python
train_data.head()
```

```python
st1="‪‪D:\Datasets\calories.csv" 
test_data = pd.read_csv(st1.strip("‪u202a"))
test_data.head()
```

```python
data = pd.concat((train_data, test_data),axis=1)
```

```python
data.head()
```

```python
data.drop('User_ID',axis=1, inplace=True)
```

```python
data.head()
```

```python
data.shape
```

```python
data.size
```

```python
data.isna().count()
```

```python
data.info()
```

```python
plt.figure(figsize=(12,7))
sbn.countplot(x="Gender",data=data,palette="spring",edgecolor = "black")
plt.title('Gender_Count')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sbn.distplot(data['Age'],hist_kws=dict(edgecolor="k", linewidth=2))
plt.show()
```

```python
plt.figure(figsize=(12,7))
sbn.histplot(data['Weight'],color='red')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sbn.boxplot(x='Gender',y='Weight',data=data)
plt.show()
```

### So we can say that male have more weight than that of women.

```python
plt.rcParams["patch.force_edgecolor"] = True
plt.figure(figsize=(12,7))
sbn.distplot(data['Height'],color="green",hist_kws=dict(edgecolor="k", linewidth=2))
plt.show()
```

```python
plt.figure(figsize=(12,7))
sbn.boxplot(x='Gender',y='Height',data=data)
plt.show()
```

### We can say that men are tall as compared to women

```python
plt.figure(figsize=(12,7))
sbn.scatterplot(x='Weight',y='Height',data=data)
plt.show()
```

### We can see here as height increases weight also increases

```python
data['BMI'] = data['Weight']/(data['Height']**2)*10000
```
