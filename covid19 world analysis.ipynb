```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize']=15,7
```

```python
country_wise=pd.read_csv(r'D:\Datasets\country_wise_latest.csv')
country_wise
```

```python
country_wise.sort_values('Confirmed',ascending=False,inplace=True)
```

```python
country_wise['index']=range(1,len(country_wise)+1)
```

```python
country_wise
```

```python
country_wise.set_index('index',inplace=True)
```

```python
country_wise['recovery_rate']=(country_wise['Recovered']/country_wise['Confirmed'])*100
country_wise['mortality_rate']=(country_wise['Deaths']/country_wise['Confirmed'])*100
country_wise
```

# NOW LETS ANALYZE THE DATA OF TOP 16 COUNTRIES


```python
filt = country_wise['Confirmed']>170000
top_16=country_wise.loc[filt,['Country/Region','Confirmed','Recovered','Active','Deaths','recovery_rate','mortality_rate']]
top_16
```

```python
top_16.style.background_gradient(cmap='Oranges')
```

```python
top_16.plot(kind='bar',x='Country/Region',y='Confirmed',color='#000000')
```

# LETS ANALYZE RECOVERY RATE AND DEATH RATE OF THE COUNTRIES


```python
plt.rcParams['figure.figsize']=15,7
plt.style.use('fivethirtyeight')
plt.bar(top_16['Country/Region'],top_16['recovery_rate'],color='#00FF00')
plt.xlabel('Country/Region')
plt.ylabel('recovery_rate(in %)')
plt.title('Recovery_rate')
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.bar(top_16['Country/Region'],top_16['mortality_rate'],color='#800000')
plt.xlabel('Country/Region')
plt.ylabel('mortality_rate(in %)')
plt.title('Mortality_rate')
plt.show()
```

```python
total_cases_in_india=india_lvl['dailyconfirmed'].sum()
print("The total recorded cases in India upto 29 of june are : ", total_cases_in_india)
```

```python
total_recovered_cases_in_india=india_lvl['dailyrecovered'].sum()
print("The total recovered cases in India upto 29 of june are : ", total_recovered_cases_in_india)
```

```python
total_deaths_in_india=india_lvl['dailydeceased'].sum()
print("The total deceased cases in India upto 29 of june are : ", total_deaths_in_india)
```

```python
state_lvl_daily=pd.read_csv(r'D:\Datasets\state_level_latest.csv')
state_lvl_daily
```

```python
state_lvl_daily.drop([0],inplace=True)
state_lvl_daily
```

```python
state_lvl_daily.loc[(1)]
```

```python
state_lvl_daily['rec_rate']=(state_lvl_daily['recovered']/state_lvl_daily['confirmed'])*100
state_lvl_daily['death_rate']=(state_lvl_daily['deaths']/state_lvl_daily['confirmed'])*100
state_lvl_daily
```

```python
req_data=state_lvl_daily.loc[:,['state','confirmed','recovered','deaths','active','rec_rate','death_rate']]
```

```python
req_data
```

```python
req_data.style.background_gradient(cmap='Reds')
```

# NOW LETS ANALYZE THE DATA OF TOP 5 STATES WITH MOST NO OF CONFIRMED NCOV 19 CASES


```python
fil=req_data['confirmed']>20000
top_5_states=req_data.loc[fil,['state','confirmed','recovered','deaths','active','rec_rate','death_rate']]
top_5_states
```

```python
plt.rcParams['figure.figsize']=15,7
top_5_states.plot(kind='bar',x='state',y=['confirmed','recovered','active','deaths'])
```
