# GOALKEEPER

```python
gk=fifa[fifa['Position']=='GK'].sort_values('Overall',ascending=False)
gk=gk[['Name','Club','Jersey Number','Nationality','Age','Overall','GKDiving','GKHandling','GKPositioning','GKReflexes','GKKicking','Value','Wage','Release Clause']]
gk
```

# LIST OF TOP 5 GOAL KEEPERS

```python
top_gk=gk.head(5)
top_gk
```

```python
top_gk['Avg']=(top_gk['GKDiving']+top_gk['GKHandling']+top_gk['GKPositioning']+top_gk['GKReflexes']+top_gk['GKKicking'])/5
```

```python
top_gk.sort_values('Avg',ascending=False,inplace=True)
```

```python
top_gk
```

```python
#NOTE :- AVG IS DIFFERENT FROM OVERALL AVG IS BASED ON PURE GOAL KEEPING SKILLS AND OVERALL CONSIDERS ALL OTHER FACTORS LIKE PASSING , SHOOTING ,ETC
```

```python
print('SO BASED ON OVERALL PERFORMANCE DE GEA LEADS AND ALSO BASED ON ONLY GOAL KEEPING SKILLS NO ONE CAN CATCH HANDS OF DE GEA ')
```

```python
print("so for ideal team we pick DE GEA  as goalkeeper and JAN OBLAK  as substitute goal keeper ")
gk_team=gk.iloc[0:2]
gk_team
```
